```python
#IMPORTING LIBRARIES
from selenium import webdriver
from time import sleep
```

```python
#CREATED EMPTY LIST FOR STORING REQUIRED DATA
Rank=[]
Crypto_Name=[]
Crypto_Symbol=[]
Crypto_MarketCap=[]
Crypto_Price=[]
Crypto_CirculatingSupply=[]
```

```python
#DEFINED A FUNCTION TO SCRAP THE REQUIRED DATA FROM WEBPAGE
def Crypto(scrape='20220109/'):
    
    #URL OF WEBPAGE
    url='https://coinmarketcap.com/historical/20220109/'
    #PATH OF SELENIUM DRIVER IN LOCAL SYSTEM
    driver=webdriver.Chrome('/home/expert/FireFox&Chrome_Selenium/chromedriver')
    driver.get(url)
    sleep(2)

    driver.maximize_window()
    
    #SCROLLING DOWN THE BODY OF THE PAGE 
    driver.execute_script("window.scrollTo(0, 2000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(2000,4000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(4000,6000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(6000,8000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(8000,10000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(10000,12000)"," ")
    sleep(30)
    
    #CLICKING THE LOAD BUTTON FOR PAGE LOADING
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    
    driver.execute_script("window.scrollTo(12000,14000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(14000,16000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(16000,18000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(18000,20000)"," ")
    sleep(30)
    
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    driver.execute_script("window.scrollTo(20000,22000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(22000,24000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(24000,26000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(26000,28000)"," ")
    sleep(30)
    
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    driver.execute_script("window.scrollTo(30000,32000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(32000,34000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(34000,36000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(36000,38000)"," ")
    sleep(30)
    
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    
    driver.execute_script("window.scrollTo(36000,38000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(38000,40000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(42000,44000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(44000,46000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(46000,48000)"," ")
    sleep(30)
    
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    driver.execute_script("window.scrollTo(48000,50000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(50000,52000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(52000,54000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(52000,56000)"," ")
    sleep(30)
    
    load_page=driver.find_element_by_xpath('.//div[@class="cmc-table-listing__loadmore"]')
    load_page.click()
    sleep(20)
    driver.execute_script("window.scrollTo(56000,58000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(58000,60000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(60000,62000)"," ")
    sleep(30)
    driver.execute_script("window.scrollTo(62000,64000)"," ")
    sleep(30)
    
    #WHOLE BODY OF THE TABLE
    tbody=driver.find_elements_by_xpath('.//div[@class="cmc-table__table-wrapper-outer"]')
    
    #EACH ROW OF THE TABLE
    tr=driver.find_elements_by_xpath('.//tr[@class="cmc-table-row"]')
    sleep(30)
    
    #RANK OF BITCOIN
    rank=driver.find_elements_by_xpath('.//td[@class="cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__rank"]')
    for r in rank:
        Rank.append(r.text.strip())

    #NAME OF BITCOIN
    name=driver.find_elements_by_xpath('.//a[@class="cmc-table__column-name--name cmc-link"]')
    for n in name:
        Crypto_Name.append(n.text.strip())
        
    #SYMBOL OF BITCOIN   
    symbol=driver.find_elements_by_xpath('.//td[@class="cmc-table__cell cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__symbol"]')[0:500]
    for s in symbol:
        Crypto_Symbol.append(s.text.strip())
        
    #MARKET CAP OF BITCOIN
    marketcap=driver.find_elements_by_xpath('.//td[@class="cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap"]')[0:500]
    for m in marketcap:
        Crypto_MarketCap.append(m.text.strip())
        
    #PRICE OF BITCOIN
    price=driver.find_elements_by_xpath('.//td[@class="cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price"]')[0:500]
    for p in price:
        Crypto_Price.append(p.text.strip())
    
    #CRYPTO CIRCULATING SUPPLY
    cirsupply=driver.find_elements_by_xpath('.//td[@class="cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply"]')[0:500]
    for c in cirsupply :
        Crypto_CirculatingSupply.append(c.text.strip())
```

```python
#RUN THE SCRAPE FUNCTION
Crypto(scrape='20220109/')
```

```python
#STORED THE DATA IN A LIST
data=list(zip(Rank,Crypto_Name,Crypto_Symbol,Crypto_MarketCap,Crypto_Price,Crypto_CirculatingSupply))
```

```python
#CREATED THE DATAFRAME
import pandas as pd
df=pd.DataFrame(data,columns=["Rank","Crypto_Name","Crypto_Symbol","Crypto_MarketCap","Crypto_Price","Crypto_CirculatingSupply"])
```
